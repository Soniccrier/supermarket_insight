```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```

```python
df.describe()
```

```python
df.drop_duplicates()
```

```python
df.groupby("branch")["total"].sum()
```

```python
branch = df.groupby('branch', as_index=False)['total'].sum()

sort = branch.sort_values('total', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'branch', y = 'total')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Total Revenue by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Revenue')
plt.show()
```

```python
branch_count = df['branch'].value_counts().reset_index()
branch_count.columns = ['branch', 'Count']

sort = branch_count.sort_values('Count', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'branch', y = 'Count')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Branch with Highest Number of Transactions')
plt.xlabel('Branch')
plt.ylabel('Transaction Count')
plt.show()
```

```python
city = df.groupby('city', as_index=False)['total'].sum()

sort = city.sort_values('total', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'city', y = 'total')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Total Revenue by Cities')
plt.xlabel('Cities')
plt.ylabel('Total Revenue')
plt.show()
```

```python
customer = df.groupby('customer_type', as_index=False)['total'].sum()

sort = customer.sort_values('total', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'customer_type', y = 'total')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Total Revenue By Customer Type')
plt.ylabel('Total Revenue')
plt.show()
```

```python
customer = df.groupby('product_line', as_index=False)['total'].sum()

sort = customer.sort_values('total', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'product_line', y = 'total')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Total Revenue from Product line')
plt.ylabel('Total Revenue')
plt.show()
```

```python
gender = df.groupby(['gender', 'product_line'], as_index=False)['total'].sum()

sort = gender.sort_values('total', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'product_line', y = 'total', hue='gender')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Total Spending by Product Line and Gender')
plt.xlabel('Product Line')
plt.ylabel('Total Revenue')
plt.show()
```

```python
df.groupby('product_line')['gross_margin_percentage'].mean().sort_values(ascending=False)
```

```python
payment = df.groupby(['payment', 'customer_type']).size().reset_index(name='Count')
payment.columns = ['payment', 'customer_type', 'Count']

sort = payment.sort_values('Count', ascending=False)

plt.figure(figsize=(12,6))
ax = sns.barplot(data=sort, x = 'payment', y = 'Count', hue='customer_type')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')

plt.title('Payment Method by Customers')
plt.xlabel('Payment Method')
plt.ylabel('Transaction Count')
plt.show()
```

```python
branch_fin = df.groupby('branch', as_index=False)[['total', 'cogs']].sum()

plt.figure(figsize=(12,6))
ax = sns.barplot(data=branch_fin, x='branch', y='total', color='green', label='Revenue')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')
ax = sns.barplot(data=branch_fin, x='branch', y='cogs', color='red', label='COGS')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')
    
plt.title('Revenue vs COGS per Branch')
plt.xlabel('Revenue')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
branch_profit = df.groupby('branch')['gross_income'].sum().reset_index()

plt.figure(figsize=(12,6))
ax = sns.barplot(data=branch_profit, x='branch', y='gross_income')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')
    
plt.title('Gross Income per Branch')
plt.xlabel('Branch')
plt.ylabel('Gross Income')
plt.show()
```

```python
price = df.groupby('date', as_index=False)['total'].sum()

plt.figure(figsize=(20,6))
sns.scatterplot(data=price, x = 'date', y = 'total', color='red', s=50)
sns.lineplot(data=price, x = 'date', y = 'total', color='blue', linestyle='--')

plt.title('Total Revenue Over Time')
plt.grid(True)
plt.show()
```

```python
price = df.groupby('date', as_index=False)['gross_income'].sum()

plt.figure(figsize=(20,6))
sns.scatterplot(data=price, x = 'date', y = 'gross_income', color='red', s=50)
sns.lineplot(data=price, x = 'date', y = 'gross_income', color='blue', linestyle='--')

plt.title('Total Income Over Time')
plt.grid(True)
plt.show()
```

```python
customer_spend = df.groupby('customer_type')['total'].sum().reset_index()

threshold = customer_spend['total'].quantile(0.75)
customer_spend['Segment'] = np.where(customer_spend['total'] >= threshold, 'High spender', 'Low spender')

customer_spend
```

```python
df.groupby('customer_type')['total'].mean()
```

```python
df.groupby('customer_type').size()
```

```python
customer_freq = df.groupby('customer_type').size().reset_index(name='Transactions')

freq_threshold = customer_freq['Transactions'].quantile(0.75)
customer_freq['FreqSegment'] = np.where(customer_freq['Transactions'] >= freq_threshold, 'High-frequency', 'Low-frequency')
customer_freq
```

```python
df_seg = df.merge(customer_spend[['customer_type','Segment']], on='customer_type', how='left')

product_pref = df_seg.groupby(['Segment','product_line'])['total'].sum().reset_index()
product_pref.sort_values(['Segment','total'], ascending=[True, False])
```

```python
basket = df.groupby(['customer_type','product_line']).size().reset_index(name='Count')
basket.sort_values(['customer_type','Count'], ascending=[True, False])

plt.figure(figsize=(12,6))
ax = sns.barplot(data=basket, x='product_line', y='Count', hue='customer_type')
for container in ax.containers:
    ax.bar_label(container, fmt='%1.0f', label_type='edge', color='black')
    
plt.title('Product Line Popularity by Customer Type')
plt.show()
```
